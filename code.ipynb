```python
df = pd.DataFrame(data)

df.replace('-', np.nan, inplace=True)
def summarize_dataframe(df):
     # 1. 缺失值统计
    missing_values = df.isnull().sum()

    # 2. 只保留有缺失值的列
    missing_values = missing_values[missing_values > 0]

    # 2. 重复值统计
    duplicate_values = df.duplicated().sum()

    # 3. 频率值最高的统计
    frequency_values = df.mode().iloc[0]  # 取每列的众数

   # 2. 只保留有缺失值的列
    missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

    # 打印结果
    print("=" * 40)
    print("缺失值统计：")
    print("=" * 40)

    if not missing_values.empty:
        for column, count in missing_values.items():
            print(f"{column}: {count} 个缺失值")
    else:
        print("没有缺失值。")

    print("=" * 40)
    print("重复值统计：")
    print("=" * 40)
    print(f"重复行的数量: {duplicate_values}")

    print("=" * 40)
    print("频率值最高的值：")
    print("=" * 40)
    for column in df.columns:
        print(f"{column}: {frequency_values[column]}")

    print("=" * 40)

# 调用函数
summarize_dataframe(df)
```

```python

# 创建图形
plt.figure(figsize=(14, 8))

# 定义颜色方案
colors = {
    'tests': ['#FF9999', '#FF3333'],
    'exams': ['#99FF99', '#33CC33']
}

# 创建气泡图
x = np.array(range(len(type_mapping)))
y_tests = avg_tests_by_type.values
y_exams = avg_exams_by_type.values

# 绘制测验次数气泡
scatter1 = plt.scatter(x - 0.15, y_tests,
                      s=y_tests * 500,
                      alpha=0.6,
                      color=colors['tests'][0],
                      edgecolors=colors['tests'][1],
                      linewidth=2,
                      label='测验次数')

# 绘制考试次数气泡
scatter2 = plt.scatter(x + 0.15, y_exams,
                      s=y_exams * 500,
                      alpha=0.6,
                      color=colors['exams'][0],
                      edgecolors=colors['exams'][1],
                      linewidth=2,
                      label='考试次数')

# 添加平均值水平线
plt.axhline(y=test_stats['mean'], color=colors['tests'][1], linestyle='--', alpha=0.3)
plt.axhline(y=exam_stats['mean'], color=colors['exams'][1], linestyle='--', alpha=0.3)

# 添加连接线
for i in range(len(x)):
    plt.plot([x[i]-0.15, x[i]+0.15], [y_tests[i], y_exams[i]],
             color='gray', linestyle='--', alpha=0.3)

# 添加数值标签
for i, (test, exam) in enumerate(zip(y_tests, y_exams)):
    plt.annotate(f'测验: {test:.2f}',
                (x[i]-0.15, test),
                xytext=(0, 10),
                textcoords='offset points',
                ha='center',
                va='bottom',
                fontsize=10)
    plt.annotate(f'考试: {exam:.2f}',
                (x[i]+0.15, exam),
                xytext=(0, 10),
                textcoords='offset points',
                ha='center',
                va='bottom',
                fontsize=10)

# 设置标题和标签
plt.title('不同类型学校的测验与考试次数比较', fontsize=14, pad=20)
plt.xlabel('学校类型', fontsize=12)
plt.ylabel('次数', fontsize=12)

# 设置x轴刻度
plt.xticks(x, [type_mapping[i] for i in range(len(type_mapping))], fontsize=10)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.3)

# 添加图例
plt.legend(fontsize=10, loc='upper right')

# 添加统计信息
stats_text = (
    f'测验统计:\n'
    f'平均值: {test_stats["mean"]:.2f}\n'
    f'最小值: {test_stats["min"]:.2f}\n'
    f'最大值: {test_stats["max"]:.2f}\n'
    f'标准差: {test_stats["std"]:.2f}\n\n'
    f'考试统计:\n'
    f'平均值: {exam_stats["mean"]:.2f}\n'
    f'最小值: {exam_stats["min"]:.2f}\n'
    f'最大值: {exam_stats["max"]:.2f}\n'
    f'标准差: {exam_stats["std"]:.2f}'
)

plt.text(0.95, 0.05, stats_text,
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'),
         fontsize=10,
         ha='right')

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()

# 打印详细统计数据
print("\n测验统计:")
print(f"平均值: {test_stats['mean']:.2f}")
print(f"最小值: {test_stats['min']:.2f}")
print(f"最大值: {test_stats['max']:.2f}")
print(f"标准差: {test_stats['std']:.2f}")

print("\n考试统计:")
print(f"平均值: {exam_stats['mean']:.2f}")
print(f"最小值: {exam_stats['min']:.2f}")
print(f"最大值: {exam_stats['max']:.2f}")
print(f"标准差: {exam_stats['std']:.2f}")

print("\n各类型学校详细数据:")
for school_type in type_mapping:
    print(f"\n{type_mapping[school_type]}:")
    print(f"测验次数: {avg_tests_by_type[school_type]:.2f}")
    print(f"考试次数: {avg_exams_by_type[school_type]:.2f}")
```

主要结论：

考试次数普遍高于测验次数，且差异较为稳定（考试标准差0.30，测验标准差0.41），表明各类学校都更重视正式考试评估
资助和私立学校的考试测验差距最大（相差约1.7-1.8次），而直资学校的考试测验频率最为接近（相差0.92次）
从三个角度分析影响：

学校竞争力角度：
资助和私立学校通过高频率正式考试来体现教学质量和竞争力
直资学校采用更均衡的评估体系，可能更注重全面发展
标准差数据显示考试频率的一致性高于测验，反映了学校间的标准化考核趋势

学生生活学习角度：
测验频率的较大标准差(0.41)表明不同学校的平时考核压力差异明显
资助和私立学校学生面临更显著的考试压力和学习节奏转换
直资学校学生能获得更稳定和连续的学习评估体验

家长选择学校角度：
可根据考测频率差异选择符合孩子学习特点的学校
关注升学成绩的家长可能更倾向选择考试频率较高的资助或私立学校
希望孩子获得均衡发展的家长可能更倾向选择直资学校的评估模式

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False     # 用来正常显示负号

# 显示图形的魔法命令
%matplotlib inline

# 示例数据
data = {
    '學校類別1': ['Gov\'t', 'Aided', 'Private', 'DSS'],
    '學校類別2': ['Whole Day', 'AM', 'PM', 'Whole Day'],
    '學生性別': ['Boys', 'Girls', 'Co-ed', 'Co-ed'],
    '教學語言': ['Chinese', 'English', 'Chinese & English', 'English'],
    '創校年份': [1950, 1960, 1970, 1980],
    'AwardsScore': [4, 3, 5, 2]
}
df = pd.DataFrame(data)

# 数据编码
encodings = {
    '學校類別1': {'Gov\'t': 1, 'Aided': 2, 'Private': 3, 'DSS': 4},
    '學校類別2': {'Whole Day': 1, 'AM': 2, 'PM': 3, 'AM+Whole Day': 4},
    '學生性別': {'Boys': 1, 'Girls': 2, 'Co-ed': 3},
    '教學語言': {'Chinese': 1, 'English': 2, 'Chinese & English': 3}
}

# 编码转换
df_encoded = df.copy()
for col, mapping in encodings.items():
    df_encoded[col] = df_encoded[col].map(mapping)

# 计算学校年龄
df_encoded['school_age'] = 2024 - df_encoded['創校年份']

# 选择要分析的列并重命名为英文（避免中文显示问题）
df_encoded = df_encoded.rename(columns={
    'AwardsScore': 'HKMO Score',
    '學校類別1': 'School Type',
    '學校類別2': 'Operation Mode',
    '學生性別': 'Gender',
    '教學語言': 'Language',
    'school_age': 'School Age'
})

cols = ['HKMO Score', 'School Type', 'Operation Mode', 'Gender', 'Language', 'School Age']
corr = df_encoded[cols].corr()

# 绘制热力图
plt.figure(figsize=(12, 10))
sns.heatmap(corr,
            annot=True,     # 显示数值
            fmt='.2f',      # 数值格式为两位小数
            cmap='RdBu_r',  # 使用红蓝配色
            vmin=-1, vmax=1,
            center=0,
            square=True,
            annot_kws={'size': 10})

plt.title('HKMO Performance Correlation Matrix', pad=20, fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
